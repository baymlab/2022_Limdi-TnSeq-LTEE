## Goal of the notebook: Identifying duplicated regions

#### Approach:
To use the output of samtools depth (saved as depth_unique.txt) to identify regions with duplications in the genome (using E.coli REL606 as our reference). Duplicated regions would have significantly higher coverage, compared to the ancestor.

#### Output: 
Text file with list of genes that have been duplicated in populations where we can detect duplications.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import time
import pandas as pd
import re
from Bio.SeqIO.FastaIO import SimpleFastaParser
import pathlib
import os
```

```python
import seaborn as sns
sns.set_style("ticks")
```

```python
#current working directory
cwd = os.getcwd()
print(cwd)
```

```python
#use the pathlib.Path function to get the parent directories-> goal is to navigate to directory with the metadata
# and the depth_unique data
path = pathlib.Path(cwd)
print(path.parents[1]) #this should be the base directory for the github repository: the exact path will differ for 
#each unique user
```

```python
metadata_path = str(path.parents[1])+'/Metadata/'
data_path = str(path.parents[1])+'/Data/WGS_Data/Sequencing_Depth/'
```

```python
#opening the pandas file with all the metadata!
all_data = pd.read_csv(metadata_path+"all_metadata_REL606.txt", sep="\t")
names = all_data.iloc[:,0]
gene_start = all_data.iloc[:,3]
gene_end = all_data.iloc[:,4]
strand = all_data.iloc[:,5]
locations = np.transpose(np.vstack([gene_start,gene_end,strand]))
uniprot_rel606 = all_data.iloc[:,6]

libraries = ['REL606', 'REL607', 'REL11330', 'REL11333', 'REL11364', 'REL11336', 'REL11339', 'REL11389', 'REL11392', 'REL11342', 'REL11345', 'REL11348', 'REL11367', 'REL11370']

#input directory with all the counts data
directory = data_path
```

```python
len_genome = 0
with open(metadata_path+"rel606_reference.fasta") as in_handle:
    for title, seq in SimpleFastaParser(in_handle):
        len_genome = len(seq)
```

extracting the coverage data

```python
coverage_all = np.zeros([len(libraries),len(seq)])
```

```python
for k in range(len(libraries)):
    fname = directory+libraries[k]+'_depth_unique.txt'
    data = pd.read_csv(fname, sep='\t',header=None)
    coverage=data.iloc[:,2].to_numpy()
    coverage_all[k,:] = coverage
    #print(data.shape) ##use this as a sanity check to see if the number of sites in depth file equals the genome length
```

## Coverage analysis

```python
average_coverage = np.mean(coverage_all,axis=1)
median_coverage = np.median(coverage_all,axis=1)
```

```python
coverage_genes = np.zeros([len(libraries),len(names)])
```

```python
for i in range(0,len(names)):
    #if gene is on the forward strand
    if locations[i,2]==1:
        coverage_genes[:,i] = np.mean(coverage_all[:,int(locations[i,0]):int(locations[i,1])],axis=1)
    elif locations[i,2]==-1:
        coverage_genes[:,i] = np.mean(coverage_all[:,int(locations[i,1]):int(locations[i,0])],axis=1)
```

```python
for k in range(1,len(libraries)):
    plt.plot(coverage_genes[0,:]/median_coverage[0], coverage_genes[k,:]/median_coverage[k], '.')
    plt.plot(coverage_genes[0,:]/median_coverage[0], coverage_genes[0,:]/median_coverage[0])
    plt.plot(coverage_genes[0,:]/median_coverage[0], coverage_genes[0,:]/median_coverage[0]*1.7)
#     plt.xlim(0,2)
#     plt.ylim(0,2)
    plt.title(f'comparing normalized coverage, ancestor and {libraries[k]}')
    plt.xlabel('Normalized coverage, REL606')
    plt.ylabel(f'Normalized coverage, {libraries[k]}')
    plt.show()
    
    plt.plot(coverage_genes[0,:]/median_coverage[0], label='ancestor')
    plt.plot(coverage_genes[k,:]/median_coverage[k], label=libraries[k])
    plt.xlabel('Gene Number')
    plt.ylabel(f'Normalized coverage')
#     plt.ylim(0,3)
    plt.legend(loc=0)
    plt.show()
```

## Examining specific populations: Ara-3, Ara-6, Ara+2, Ara+5

These populations appear to have large duplications. I will zoom into the interesting parts of the genomes for these specific populations

### Ara-6

```python
span_left = 2000
span_right = 2500
k = 7
genes = np.linspace(span_left,span_right-1,span_right-span_left)
plt.plot(genes, coverage_genes[k,span_left:span_right]/median_coverage[k]/coverage_genes[0,span_left:span_right]*median_coverage[0], label=libraries[k])
plt.xlabel('Gene Number')
plt.ylabel('Relative Coverage (normalized to ancestor)')
plt.legend(loc=0)
plt.title('Large duplication spanning ~300 genes in REL11389')
```

```python
# identifying duplicated region:
large = np.where(coverage_genes[k,span_left:span_right]/median_coverage[k]/coverage_genes[0,span_left:span_right]*median_coverage[0]>1.35)[0]+span_left
```

```python
span_left = 3550
span_right = 3650
k = 7
genes = np.linspace(span_left,span_right-1,span_right-span_left)
plt.plot(genes, coverage_genes[k,span_left:span_right]/median_coverage[k]/coverage_genes[0,span_left:span_right]*median_coverage[0], label=libraries[k])
plt.xlabel('Gene Number')
plt.ylabel('Relative Coverage (normalized to ancestor)')
plt.legend(loc=0)
plt.title('Large duplication spanning ~300 genes in REL11389')
```

```python
# identifying start and end of duplications
small = np.where(coverage_genes[k,span_left:span_right]/median_coverage[k]/coverage_genes[0,span_left:span_right]*median_coverage[0]>1.5)[0]+span_left
```

```python
# list of all genes that are duplicated in Ara-6
dup_389 = np.concatenate([large, small])
```
